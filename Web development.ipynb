 1.  Explain the main components of client-server architecture and their functions.
   -  The main components of client-server architecture are clients, servers, and networks. The client-server model is a network architecture that distributes tasks between clients and servers.
   # Components
   -  Clients : Devices like computers, tablets, or smartphones that request services or resources from servers
Servers : Powerful computers or specialized devices that respond to client requests
Networks : The connection between clients and servers, often using the TCP/IP protocol suite .
